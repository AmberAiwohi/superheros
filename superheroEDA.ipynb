```python
# Load needed packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import regex as re
```

```python
# Load dataframe
df = pd.read_csv('movies.csv')
df.head(10)
```

```python
# take out commas in gross value
df['Gross'] = df['Gross'].replace(",", "",regex=True)
df.head(10)
```

```python
# change gross to numeric
df['Gross'] = pd.to_numeric(df['Gross'])
df.head(10)
```

```python
# take out min unit market from runtime
df['Runtime'] = df['Runtime'].replace(" min", "",regex=True)
df.head(10)
```

```python
# change runtime to numeric
df['Runtime'] = pd.to_numeric(df['Runtime'])
df.head(10)
```

```python
# Clean year column
df['Year'] = df['Year'].str.extract('(\d+)', expand=False)
```

```python
# scatter plot of runtime vs. gross
runtime = px.scatter(data_frame = df, x = 'Runtime', y = 'Gross', 
           hover_data=df.columns, 
           color = 'Rating',
           trendline="lowess",
           title = 'Runtime vs. Gross',
           labels={"Runtime": "Runtime (min)",
                 }) 
runtime.write_html('runtime.html')
runtime
```

```python
# scatter plot of runtime vs. gross with lowess line
p = sns.lmplot(data = df, x="Runtime", y="Gross", lowess=True).set(title='Runtime vs. Gross w/ Lowess Line');
p.set_axis_labels( "Runtime (min)" , "Gross (million)");
p.savefig('runtime_lowess.png')
```

```python
# boxplot of rating vs. gross
sns.boxplot(data = df, x = 'Rating', y = 'Gross', palette = 'Blues').set(title='Rating vs. Gross');
plt.xticks(rotation=45);
plt.ylabel('Gross (million)');
plt.savefig('rating.png')
```

```python
# create year df of value counts per year
year_df = pd.DataFrame(df[['Year']].value_counts())
year_df = year_df.reset_index()
```

```python
# barplot of num of movies produced per year
year_df.plot(x = 'Year', y = 0,kind='bar');
plt.title('Number of Superhero Movies Produced Per Year')
plt.ylabel('Count');
plt.savefig('mov_per_year')
```
