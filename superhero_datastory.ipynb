```python
# Load needed packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import regex as re
```

```python
# Load dataframe
df = pd.read_csv('movies.csv')
df.head(10)
```

```python
# take out commas in gross value
df['Gross'] = df['Gross'].replace(",", "",regex=True)
df.head(10)
```

```python
# change gross to numeric
df['Gross'] = pd.to_numeric(df['Gross'])
df.head(10)
```

```python
# take out min unit market from runtime
df['Runtime'] = df['Runtime'].replace(" min", "",regex=True)
df.head(10)
```

```python
# change runtime to numeric
df['Runtime'] = pd.to_numeric(df['Runtime'])
df.head(10)
```

```python
# Clean year column
df['Year'] = df['Year'].str.extract('(\d+)', expand=False)
```

```python
# take out \n 
df['Genre'] = df['Genre'].replace("\n", "",regex=True)
df
```

```python
# change year to numeric
df['Year'] = pd.to_numeric(df['Year'])
df.head(10)
```

```python
# data story plot 1 - runtime and rating 
plot_1 = px.scatter(data_frame = df, x = 'Runtime', y = 'Gross', 
           hover_data=df.columns, 
           color = 'Rating',
           trendline="lowess",
           title = 'Exploring the Runtime and Rating of Superhero Movies',
           labels={"Runtime": "Runtime (min)",
                 }) 
plot_1.write_html('plot_1.html')
plot_1
```

```python
# data story plot 1 - year and genre
plot_2 = px.scatter(data_frame = df, x = 'Year', y = 'Gross', 
           hover_data=df.columns, 
           trendline="lowess",
           color = 'Genre',
           title = 'Exploring the Year and Genre of Superhero Movies'
           ) 
plot_2.write_html('plot_2.html')
plot_2
```
